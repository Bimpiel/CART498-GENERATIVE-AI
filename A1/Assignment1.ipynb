```python
import torch
from diffusers import StableDiffusionPipeline

# Load the model
model_id = "CompVis/stable-diffusion-v1-4"
device = "cuda"

pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to(device)

# Define the prompts and number of images per prompt
prompts = [
    "What do the stars look like from the sky? make it sureal",
    "A vibrant nebula floating in deep space make it sureal",
    "A surreal view of galaxies colliding in space make it sureal"
]
num_images = 1

# Generate images for each prompt
for prompt_idx, prompt in enumerate(prompts, 1):
    images = pipe(prompt, num_images_per_prompt=num_images).images
    for i, image in enumerate(images):
        image.save(f"Prompt_{prompt_idx}_Image_{i+1}.png")
```
